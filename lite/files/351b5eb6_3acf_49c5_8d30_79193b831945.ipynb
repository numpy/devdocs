```python
import numpy as np
a = np.ma.array([1.0, 2.0, 3.0], mask=[0, 1, 0])
a
```

Use ``fill_value`` to set a custom fill value on the view without
copying the data:


```python
a.view(fill_value=-999.0)
```

View as a plain :class:`numpy.ndarray` — the mask is not preserved:


```python
a.view(np.ndarray)
```
