```python
import numpy as np
```

Sum each column, skipping masked values:


```python
a = np.ma.array([[1, 2, 3],
                 [4, 5, 6]], mask=[[0, 1, 0],
                                   [0, 0, 1]])
np.ma.apply_along_axis(np.ma.sum, 0, a)
```

Compute the mean of each row, ignoring masked values:


```python
np.ma.apply_along_axis(np.ma.mean, 1, a)
```
